```python

response = requests.post(TOKEN_URL, data=auth_data)  # Send a POST request to get an access token.
access_token = response.json()["access_token"]  # Extract the access token from the response.

# Step 2: Define headers with authentication for further requests
headers = {"Authorization": f"Bearer {access_token}"}  # Include the token in the Authorization header.

# Step 3: Fetch adoptable pets using Petfinder API
PET_URL = "https://api.petfinder.com/v2/animals"  # API endpoint to retrieve pet listings.

params = {
    "type": "dog",  # Specify the type of pet (e.g., "dog" or "cat").
    "location": "75001",  # Filter results by location (ZIP code for Paris).
      # Limit the number of results to 5 pets.
}

response = requests.get(PET_URL, params=params, headers=headers)  # Make a GET request to retrieve pet listings.

# Step 4: Check if the API request was successful
if response.status_code == 200:  # If the status code is 200, it means the request was successful.
    pet_data = response.json()  # Convert the response to JSON format.
    
    # Step 5: Iterate through the list of pets and display details
    for pet in pet_data["animals"]:
        print(f"üê∂ Name: {pet['name']}")  # Print the pet's name.
        print(f"üìå Breed: {pet['breeds']['primary']}")  # Print the pet's primary breed.
        print(f"üéÇ Age: {pet['age']}")  # Print the pet's age category.
        print(f"üìç Location: {pet['contact']['address']['city']}, {pet['contact']['address']['state']}")  # Print the city and state of the shelter.
        print(f"üè† Shelter: {pet['organization_id']}")  # Print the shelter's unique ID.
        print(f"üîó Adoption Link: {pet['url']}")  # Print the link to the pet's adoption page.
        print("-" * 40)  # Print a separator for readability.

else:  # If the request fails, display an error message.
    print(f"‚ùå Failed to fetch pets: {response.status_code}")  # Print the HTTP status code for debugging.
```

```python

# Exemple : obtenir les donn√©es des animaux
PET_URL = "https://api.petfinder.com/v2/animals"
params = {
    "type": "dog",
    "location": "75001",
    "limit": 100
}
response = requests.get(PET_URL, params=params, headers=headers)

if response.status_code == 200:
    data = response.json()         # La r√©ponse enti√®re en JSON
    pet_list = data["animals"]       # La liste des animaux (c'est d√©j√† une liste de dictionnaires)
    
    # Utilise json_normalize pour aplatir les donn√©es imbriqu√©es :
    df = pd.json_normalize(pet_list, sep='_')
    
    # Affiche un aper√ßu du DataFrame
    df.head()
else:
    print(f"‚ùå Failed to fetch pets: {response.status_code}")
df.head()

df.to_excel("adoptable_pets.xlsx", index=False, sheet_name="Pets")
df.to_csv("adoptable_pets.csv", index=False)
df.shape

```

```python


auth_data = {"grant_type": "client_credentials", "client_id": CLIENT_ID, "client_secret": CLIENT_SECRET}
response = requests.post(TOKEN_URL, data=auth_data)
access_token = response.json()["access_token"]

# Define headers with authentication
headers = {"Authorization": f"Bearer {access_token}"}

# üê∂ Fetch adoptable pets from Petfinder API
PET_URL = "https://api.petfinder.com/v2/animals"
params = {"type": "dog", "location": "75001", "limit": 5}  # Example: Dogs near Paris

response = requests.get(PET_URL, params=params, headers=headers)

if response.status_code == 200:
    pet_data = response.json()["animals"]  # Extract list of pets

    # üìä Convert JSON response directly into DataFrame
    df_pets = pd.json_normalize(pet_data)

    # üèó Select and rename relevant columns
    df_pets_selected = df_pets[["name", "breeds.primary", "age", "organization_id", "url", "contact.address.city", "contact.address.state"]]
    df_pets_selected.columns = ["Name", "Breed", "Age", "Shelter ID", "Adoption Link", "City", "State"]

    print("üê∂ Adoptable Pets DataFrame:")
    print(df_pets_selected.head())  # Display a preview of the pet listings

else:
    print(f"‚ùå Failed to fetch pets: {response.status_code}")

# üè† Fetch shelter details (example for one shelter)
shelter_id = "TX2896"  # Example shelter ID
ORG_URL = f"https://api.petfinder.com/v2/organizations/{shelter_id}"

response = requests.get(ORG_URL, headers=headers)
if response.status_code == 200:
    shelter_data = response.json()["organization"]

    # Convert shelter data to DataFrame
    df_shelters = pd.DataFrame([{
        "Shelter ID": shelter_data["id"],
        "Shelter Name": shelter_data["name"],
        "City": shelter_data["address"]["city"],
        "State": shelter_data["address"]["state"],
        "Phone": shelter_data["phone"],
        "Website": shelter_data["url"]
    }])

    print("\nüè† Shelter Details DataFrame:")
    print(df_shelters.head())  # Display shelter details

else:
    print(f"‚ùå Failed to fetch shelter details: {response.status_code}")

# üîó Merge Pet & Shelter DataFrames
df_combined = df_pets_selected.merge(df_shelters, on="Shelter ID", how="left")
print("\nüîÑ Merged DataFrame:")
df_combined.head()  # Display final merged data
```
